```python
path_to_zip_file = "Kickstartercombo.zip"
directory_to_extract_to = ""

import zipfile
zip_ref = zipfile.ZipFile(path_to_zip_file, 'r')
zip_ref.extractall(directory_to_extract_to)
zip_ref.close() 
```

```python
import pandas as pd
Location = "Kickstartercombo.csv"
df = pd.read_csv(Location)
df.head()
```

```python
df.columns = ['One', 'ID', 'Name', 'Main_Category', 'Category', 'Country', 'State', 'Goal', 'Pledged', 'Currency', 'Launched', 'Deadline', 'Backers', 'USD_Pledged', 'USD_Goal_Real', 'USD_Pledged_Real']
df.head()
```

```python
#code to take care of missing data in USD_Goal_Real
import numpy as np
df['USD_Goal_Real2'] = np.where((pd.isnull(df['USD_Goal_Real']))&(df['Currency']=='USD'), df['Goal'], df['USD_Goal_Real'])
df.head()
```

```python
#code to take care of missing data in USD_Pledged_Real
df['USD_Pledged_Real2'] = np.where((pd.isnull(df['USD_Pledged_Real']))&(df['Currency']=='USD'), df['Pledged'], df['USD_Pledged_Real'])
df.head()
```

```python
df['State2'] = np.where(df['State']=="successful", 1, 0)
df['Currency2'] = np.where(df['Currency']=='USD', 1, 0)
df.head()
```

```python
pd.value_counts(df['State'])
```

```python
pd.value_counts(df['State2'])
```

```python
df_currency = pd.get_dummies(df['Currency2'])
df_currency.head()
```

```python
df = df.join(df_currency)
df.tail()
```

```python
headers = list(df.columns.values)
```

```python
headers[20]= 'Non_USD'
headers[21] = 'USD'
df.columns = headers
df.head()
```

```python
df.describe()
```
